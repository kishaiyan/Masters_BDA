```python
df_before=pd.read_csv("Mental Health 2018-2021/before_covid.csv")
```

```python
columns_to_drop=['EDUC','ETHNIC','RACE','GENDER','SPHSERVICE','CMPSERVICE','OPISERVICE','RTCSERVICE','IJSSERVICE','MH1','MH2','MH3','SUB','MARSTAT','SMISED','SAP','EMPLOY','DETNLF','LIVARAG','STATEFIP','DIVISION','REGION','CASEID']
```

```python
df_before=df_before.drop(columns=columns_to_drop)
```

```python
df_before.shape
```

```python
column_name='VETERAN'
```

```python
df_before[column_name] = np.where(df_before[column_name] == 2, 1,df_before[column_name])
```

```python
df_before[column_name]= np.where(df_before[column_name]< 0,0,df_before[column_name])
```

```python
import matplotlib.pyplot as plt

# Assuming df_before_cleaned is your DataFrame and 'ANXIETYFLG' and 'AGE' are columns
# Filter and sort df_before_cleaned to create df_anxiety
df_anxiety = df_before_cleaned[df_before_cleaned['ANXIETYFLG'] == 1]
df_anxiety = df_anxiety.sort_values(by='AGE')

# Group by 'AGE' and count occurrences of 'ANXIETYFLG'
age_counts = df_anxiety.groupby('AGE')['ANXIETYFLG'].count().reset_index()

# Plotting a bar plot for age counts
plt.figure(figsize=(12, 8))
plt.bar(age_counts['AGE'], age_counts['ANXIETYFLG'], color='b', alpha=0.7)
plt.title('Count of Anxiety by Age')
plt.xlabel('Age')
plt.ylabel('Count of Anxiety')
plt.grid(True)
plt.show()
```

```python
df_anxiety = df_after_cleaned[df_after_cleaned['ANXIETYFLG'] == 1]
df_anxiety = df_anxiety.sort_values(by='AGE')

# Group by 'AGE' and count occurrences of 'ANXIETYFLG'
age_counts = df_anxiety.groupby('AGE')['ANXIETYFLG'].count().reset_index()

# Plotting a bar plot for age counts
plt.figure(figsize=(12, 8))
plt.bar(age_counts['AGE'], age_counts['ANXIETYFLG'], color='r', alpha=0.7)
plt.title('Count of Anxiety by Age')
plt.xlabel('Age')
plt.ylabel('Count of Anxiety')
plt.grid(True)
plt.show()
```
